```python
from qiskit import *
from qiskit import Aer
from qiskit.visualization import plot_state_city, plot_histogram
```

création d'un circuit avec 2 qbit, un par gardiens + 1 pour le mensonge1

```python
circ = QuantumCircuit(3)
```

ajout d'une porte de Hadamar, pour mettre le trésors de matière aléatoire

```python
circ.h(0)
```

ajout d'un porte cnot pour intriquer les 2 qbit

```python
circ.cx(0,1)
```

```python
circ.h(2)
```

parcour mensonge

```python
circ.cx(2, 1)
circ.x(2, "not")
circ.cx(2,0)
circ.x(2, "not")
```

question: quel porte l'autre gardiens me dirai de prendre

```python
circ.swap(0,1)
```

```python
circ.x(0, "not")
circ.x(1, "not")
```

```python
circ.cx(2, 1)
circ.x(2, "not")
circ.cx(2,0)
circ.x(2, "not")
```

mesures des qbits

```python
circ.measure_all()
```

affichage du circuit

```python
circ.draw("mpl")
```

Execusion & Vizualisation

```python
simulator = Aer.get_backend('aer_simulator')
```

```python
circ = transpile(circ, simulator)
result = simulator.run(circ).result()
counts = result.get_counts(circ)
plot_histogram(counts, title='Bell-State counts')
```
